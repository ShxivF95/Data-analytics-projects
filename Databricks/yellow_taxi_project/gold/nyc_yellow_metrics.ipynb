```python
%sql
CREATE TABLE IF NOT EXISTS nyc_yellow.gold.daily_trip_metrics
USING DELTA
AS
SELECT
  date(tpep_pickup_datetime) AS trip_date,
  COUNT(*) AS total_trips,
  SUM(total_amount) AS total_revenue
FROM nyc_yellow.silver.nyc_yellow_silver
GROUP BY date(tpep_pickup_datetime);
```
